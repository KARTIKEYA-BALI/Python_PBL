```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,precision_score,classification_report,f1_score,recall_score,confusion_matrix
from sklearn.model_selection import train_test_split
```

```python

x=data['vector_space']
y=data['label']
```

```python
"""THIS IS FOR 20 Percent TESTING DATA"""

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from scipy import sparse
"""LINEAR SVM"""


x_trained_sparse_data=sparse.vstack(x_train.tolist())
x_test_sparse_data=sparse.vstack(x_test.tolist())
svm_linear=SVC(kernel='linear')
svm_linear.fit(x_trained_sparse_data,y_train)
y_pred_linear=svm_linear.predict(x_test_sparse_data)
```

```python
cm_linear = confusion_matrix(y_test, y_pred_linear)
recall_linear=recall_score(y_test,y_pred_linear,average='weighted')
accuracy_linear=accuracy_score(y_test,y_pred_linear)
f1_linear=f1_score(y_test,y_pred_linear)
```

```python
print("FOR LINEAR\n")
print(f"confusion matrix:\n",cm_linear)
print("recall score is:",recall_linear)
print("accuracy_score is:",accuracy_linear)
print("f1 score is :",f1_linear)
```

```python
"""POLYNOMIAL SVM WITH DEGREE 3"""
svm_poly_3=SVC(kernel='poly',degree=3)
svm_poly_3.fit(x_trained_sparse_data,y_train)
y_pred_poly_3=svm_poly_3.predict(x_test_sparse_data)
```

```python
cm_poly_3 = confusion_matrix(y_test, y_pred_poly_3)
recall_poly_3=recall_score(y_test,y_pred_poly_3,average='weighted')
accuracy_poly_3=accuracy_score(y_test,y_pred_poly_3)
f1_poly_3=f1_score(y_test,y_pred_poly_3)
cls_rep_poly_3=classification_report(y_test,y_pred_poly_3)
```

```python
print("FOR POLYNOMIAL WITH DEGREE 3 \n")
print(f"confusion matrix:\n",cm_poly_3)
print("recall score is:",recall_poly_3)
print("accuracy_score is:",accuracy_poly_3)
print("f1 score is :",f1_poly_3)
print("\nclassification report is:\n",cls_rep_poly_3)
```
